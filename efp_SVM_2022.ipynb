```
#SVM COM BALANCEAMENTO

import pandas as pd 
from sklearn.model_selection import KFold, StratifiedKFold, StratifiedShuffleSplit, cross_val_score, cross_val_predict, GridSearchCV
import numpy as np
from sklearn import preprocessing
from sklearn import decomposition
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE

seed = 42
best = 0
num_folders = 10;

timenow()
for param in range(1,11):
  df = sample_df
  X = df.drop([className], axis=1).values
  Y = df[className].values

  feature_names = list(df.drop([className], axis=1).columns.values)
  target_name = [className]

  processed_X = pd.DataFrame(X, columns=feature_names) #DataFrame sem a classe (criação de DataFrame dos dados com os nomes das colunas)
  folder = StratifiedKFold(n_splits=num_folders, random_state=seed, shuffle=True) #objeto folder para separar CV-10

  kernel = 'rbf'

  cost =  0#input
  gamma = 0#input

  f1_cvscores = []
  f1_mean = []
  f1_std = []

  recall_cvscores = []
  recall_mean = []
  recall_std = []

  precision_cvscores = []
  precision_mean = []
  precision_std = []

  print(key)
  print(cost)
  print(gamma)
  fig_count = 1;
  
  for index, (train, test) in enumerate(folder.split(X, Y)):
      scaler = MinMaxScaler(feature_range=(0,1))
      s_train = scaler.fit_transform(X[train])
      (X_train, Y_train) = (s_train, Y[train])

      #(X_train, Y_train) = (X[train], Y[train])

      print("Training on fold " + str(index+1) + "/10...")
      #Balanceamento do conjunto de treinamento (em um i-fold especificamente)    
      #----------smote normal----------------------------
      sm = SMOTE(sampling_strategy='not majority', random_state=seed, n_jobs=-1, k_neighbors=param)
      #import pdb; pdb.set_trace()
      X_train, Y_train = sm.fit_resample(X_train, Y_train)
      #import pdb; pdb.set_trace()  
      #-------------svm-smote------------------------------
      #from imblearn.over_sampling import SVMSMOTE
      #model = SVC(C=cost, gamma=gamma, kernel = kernel, random_state=seed)
      #svmST = SVMSMOTE(sampling_strategy='not majority', svm_estimator=model, random_state=seed, n_jobs=-1, k_neighbors=param)
      #(X_train, Y_train) = svmST.fit_resample(s_train, Y[train])
      #-------------borderline-smote------------------------------
      #from imblearn.over_sampling import BorderlineSMOTE
      #borderSmote = BorderlineSMOTE(sampling_strategy='not majority', random_state=seed, n_jobs=2)
      #(X_train, Y_train) = borderSmote.fit_resample(X[train], Y[train])
      #//////////////////TOMEK////////////////////////////
      #sampler_classes = list(["'OXIDORREDUTASES '", "TRANSFERASES", "HIDROLASES"])
      #from imblearn.under_sampling import TomekLinks
      #tl = TomekLinks(sampling_strategy=sampler_classes)
      #(X_train,Y_train) = tl.fit_sample(X_train, Y_train)
      #//////////////////////////////////////////////////
      
      model = SVC(C=cost, gamma=gamma, kernel = kernel, random_state=seed)

      #Treinamento
      model.fit(X_train, Y_train)

      #Teste
      s_test = scaler.transform(X[test])
      y_pred = model.predict(s_test)

      #Avaliação
        
      f1_micro = f1_score(Y[test], y_pred, average='micro')  
      f1_macro = f1_score(Y[test], y_pred, average='macro')    
      
      f1_cvscores.append(f1_score(Y[test], y_pred, average=None))
      recall_cvscores.append(recall_score(Y[test], y_pred, average=None))
      precision_cvscores.append(precision_score(Y[test], y_pred, average=None))

      cm = confusion_matrix(Y[test], y_pred, labels=model.classes_)
      print(cm)
      disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
      import matplotlib.pyplot as plt
      disp.plot()
      tick_marks = np.arange(len(model.classes_))
      plt.xticks(tick_marks, ['EC1','EC3','EC5','EC4','EC6','EC2'], rotation=0)
      plt.yticks(tick_marks, ["Oxidoreductases (EC1)",'Hydrolases (EC3)','Isomerases (EC5)','Lyases (EC4)','Ligases (EC6)','Transferases (EC2)'],  rotation=0)
      plt.savefig('/content/cm_folder'+ str(fig_count)+'.png', format='png', dpi=300, bbox_inches = 'tight')
      fig_count = fig_count+1
     
      plt.show()  
      print(classification_report(Y[test], y_pred, digits=8))
 
#To-do: refactoring code below to void duplicating code.
  f1_values = list(zip(*f1_cvscores))
  recall_values = list(zip(*recall_cvscores))
  precision_values = list(zip(*precision_cvscores))
  
  
    #for i, val in enumerate(f1_values):
  for i in range(0, numClasses):
      f1_mean.append(np.mean(f1_values[i]))
      f1_std.append(np.std(f1_values[i]))
      recall_mean.append(np.mean(recall_values[i]))
      recall_std.append(np.std(recall_values[i]))
      precision_mean.append(np.mean(precision_values[i]))
      precision_std.append(np.std(precision_values[i]))
 
  #import pdb; pdb.set_trace()
  print("-----\nMédia por Classe (F-measure):")
  print('(<valores da classe em questão>)\n')
  for x in range(0, numClasses):
      print("%s: %.4f (+/- %.4f)" % (model.classes_[x], f1_mean[x], f1_std[x]))
      print(f1_values[x],'\n')
  print("%s: %.4f (+/- %.4f)" % ('Average', np.mean(f1_cvscores), np.std(f1_cvscores)))

  print("-----\nMédia por Classe (Recall): \n")
  print('(<valores da classe em questão>)\n')
  for x in range(0, numClasses):
      print("%s: %.4f (+/- %.4f)" % (model.classes_[x], recall_mean[x], recall_std[x]))
      print(recall_values[x],'\n')
  print("%s: %.4f (+/- %.4f)" % ('Average', np.mean(recall_cvscores), np.std(recall_cvscores)))

  print("-----\nMédia por Classe (Precision): \n")
  print('(<valores da classe em questão>)\n')
  for x in range(0, numClasses):
      print("%s: %.4f (+/- %.4f)" % (model.classes_[x], precision_mean[x], precision_std[x]))
      print(precision_values[x],'\n')
  print("%s: %.4f (+/- %.4f)" % ('Average', np.mean(precision_cvscores), np.std(precision_cvscores)))
  

  print('==========================================================')  
  print('param_K = ', param)
  if np.mean(f1_cvscores)>best:
    best = np.mean(f1_cvscores)
  #import pdb;pdb.set_trace()
print(best)
timenow()

 #------------------------------------------------------
 #OTHER IMPLEMENTATIONS
 #-------------------------------------------------------
 #//////////////////SMOTE-TOMEK////////////////////////////
  #from imblearn.combine import SMOTETomek
  # stl = SMOTETomek(sampling_strategy='not majority', random_state=seed)
  #(X_train,Y_train) = stl.fit_sample(X_train, Y_train)

 #///////////////////////ADASYN////////////////////////////
   #from imblearn.over_sampling import ADASYN
   #ada = ADASYN(sampling_strategy='not majority', random_state=seed, n_jobs=-1)
   #(X_train, Y_train) = ada.fit_resample(X[train], Y[train])
```
